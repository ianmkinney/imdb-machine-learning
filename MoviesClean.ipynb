```python
import pandas as pd
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor
```

```python
# The path to our CSV file
file = "Resources/IMDb movies.csv"

# Read our Kickstarter data into pandas
movies_df = pd.read_csv(file)
movies_df.head()
```

```python
# The path to our CSV file
file = "Resources/IMDb ratings.csv"

# Read our Kickstarter data into pandas
ratings_df = pd.read_csv(file)
ratings_df
```

```python
merged_df = pd.merge(movies_df, ratings_df, on='imdb_title_id')
merged_df
```

```python
merged_dropped_df = merged_df.dropna(axis=0)
merged_dropped_df = merged_dropped_df[merged_dropped_df['country'].str.contains('USA')]
merged_dropped_df = merged_dropped_df[merged_dropped_df['year'].astype(int) > 2000]
merged_dropped_df
```

```python
len(merged_dropped_df['genre'].unique())

genres = []

for genre in merged_dropped_df['genre'].unique():
    values = genre.split(',')
    for value in values:
        if value in genres:
            break
        else:
            genres.append(value.replace(' ', ''))

for index, row in merged_dropped_df.iterrows():
    for genre in genres:
        if genre in row['genre']:
            merged_dropped_df[index, genre] = 1
        else:
            merged_dropped_df[index, genre] = 0

merged_dropped_df
```

```python
# Add gender columns

merged_dropped_df['genre'].unique()
```

```python
randomized_merged_dropped_df = merged_dropped_df.sample(n=10000, replace=True).reset_index(drop=True)
randomized_merged_dropped_df.head()
```

```python
randomized_merged_dropped_df.to_csv('Sample/sample.csv')
```

```python
# The path to our CSV file
file = "Sample/sample.csv"

# Read our Kickstarter data into pandas
sample_df = pd.read_csv(file)
sample_df.head()
```

```python
sample_df.columns
```

```python
sample_df[['weighted_average_vote','mean_vote','avg_vote']].corr()
```

```python
y=sample_df['us_voters_rating']
X=sample_df[['year','duration','weighted_average_vote']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Scaling the X data by using StandardScaler()
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_train_scaled
```

```python
# Transforming the test dataset based on the fit from the training dataset
X_test_scaled = scaler.transform(X_test)
X_test_scaled
```

```python
# Create the model

from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# Fit the model to the training data. 

### BEGIN SOLUTION

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Use our model to make predictions
predicted = model.predict(X_test)

# Score the predictions with mse and r2
mse = mean_squared_error(y_test, predicted)
r2 = r2_score(y_test, predicted)

print(f"mean squared error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```
