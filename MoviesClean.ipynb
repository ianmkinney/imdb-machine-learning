```python
import pandas as pd
```

```python
# The path to our CSV file
file = "Resources/IMDb movies.csv"

# Read our Kickstarter data into pandas
movies_df = pd.read_csv(file)
movies_df.head()
```

```python
# The path to our CSV file
file = "Resources/IMDb ratings.csv"

# Read our Kickstarter data into pandas
ratings_df = pd.read_csv(file)
ratings_df.head()
```

```python
merged_df = pd.merge(movies_df, ratings_df, on='imdb_title_id')
merged_df
```

```python
merged_dropped_df = merged_df.dropna(axis=1)
merged_dropped_df
```

```python
randomized_merged_dropped_df = merged_dropped_df.sample(n=10000)
randomized_merged_dropped_df
```
