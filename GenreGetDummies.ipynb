```python
%matplotlib inline
import pandas as pd
import numpy as np

# The path to our CSV file
file = "Sample/sample.csv"

# Read our data into pandas
sample_df = pd.read_csv(file)
sample_df.drop('Unnamed: 0', axis=1, inplace=True)
sample_df
```

```python
# Get rid of duplicate movies using the imdb_title_id column
sample_nodupes_df = sample_df.drop_duplicates(subset=['imdb_title_id'], keep='last')
sample_nodupes_df
```

```python
# Print value count of original titles to look further into duplicate movies
sample_nodupes_df['original_title'].value_counts()
```

```python
# Print test title from original_title column to ensure they are not duplicates
sample_nodupes_df.loc[sample_nodupes_df['original_title']=='Home']
```

```python
# Split up genres and stack them by original title
test_df = pd.DataFrame(sample_nodupes_df.genre.str.split(', ').tolist(), index=sample_nodupes_df.original_title).stack()
test_df
```

```python
# Reset the index
test_df = test_df.reset_index([0,'original_title'])
test_df
```

```python
# Get dummies for genre column
dummy = pd.get_dummies(test_df, columns=[0])
dummy
```

```python
# Groupby original title column and get max count of genres associated to each title
genre_df = dummy.groupby(['original_title']).max().reset_index()
genre_df
```

```python
merged_df = pd.merge(sample_df, genre_df, on='original_title')
merged_df.head()
```

```python
merged_df.to_csv('Sample/learning_sample.csv')
```

```python
merged_df['date_published'] = merged_df['date_published'].str.replace('-','')
merged_df.head()
```

```python
merged_df['avg_vote_round'] = merged_df['avg_vote'].astype(int).round(0)
merged_df.head()
```

```python
# Learing TEST

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


model = LinearRegression()

y=merged_df['avg_vote_round']
X=merged_df[['year', 'actor_number', 'production_number', 'director_number', 'writer_number', 'budget', 'duration', '0_Action', '0_Adventure', '0_Animation',
       '0_Biography', '0_Comedy', '0_Crime', '0_Drama', '0_Family',
       '0_Fantasy', '0_History', '0_Horror', '0_Music', '0_Musical',
       '0_Mystery', '0_News', '0_Romance', '0_Sci-Fi', '0_Sport', '0_Thriller',
       '0_War', '0_Western']]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_train_scaled

X_test_scaled = scaler.transform(X_test)
X_test_scaled

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Use our model to make predictions
predicted = model.predict(X_test)

# Score the predictions with mse and r2
mse = mean_squared_error(y_test, predicted)
r2 = r2_score(y_test, predicted)

print(f"mean squared error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train_scaled, y_train)
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```

```python
feature_importances = clf.feature_importances_
```

```python
features = sorted(zip(X.columns, clf.feature_importances_), key = lambda x: x[1])
cols = [f[0] for f in features]
width = [f[1] for f in features]

fig, ax = plt.subplots()

fig.set_size_inches(10,200)
plt.margins(y=0.001)

ax.barh(y=cols, width=width)

plt.show()
```
